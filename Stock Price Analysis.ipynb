```python
from collections import defaultdict
import datetime as dt

import numpy as np
import matplotlib.pyplot as plt

import networkx as nx
import pandas

from __future__ import division
```

### Load and clean Enron data

```python
data = pandas.read_pickle("AllEmails.pickle")
```

```python
data["Year"] = data.Date.apply(lambda x: x.year)
data["Month"] = data.Date.apply(lambda x: x.month)
data["Day"] = data.Date.apply(lambda x: x.day)
```

```python
# Subset on dates
data = data[(data.Year >= 1999) & (data.Year <= 2002)]
```

```python
# Refactor columns
data.rename(columns={"Date": "FullDate"}, inplace=True)
data["Date"] = data.FullDate.apply(lambda x: pandas.Timestamp(x.date()))
```

### Load and clean Enron stock data

```python
enron_stock = pandas.read_csv("EnronStock.csv", parse_dates=[0])
```

```python
enron_close = enron_stock[["Date", "Close"]].set_index("Date")["Close"]
```

```python
enron_close.plot()
```

```python
stock_delta = enron_close.diff()
```

```python
stock_delta.plot()
```

## Stock movement vs email volume

Does the overall volume of all emails sent daily correspond to the stock movement?

```python
# Daily volume
daily_volume = data.groupby("Date").aggregate(len)["From"]
```

```python
daily_volume.plot()
```

```python
stock_delta.corr(daily_volume)
```

```python
daily_volume.corr(stock_delta.abs())
```

```python
df = pandas.DataFrame({"Emails": daily_volume, "Close": enron_close})
df = df[pandas.notnull(df.Emails) & pandas.notnull(df.Close)]
df["PriceChange"] = df.Close.diff()
```

```python
df["Volatility"] = pandas.rolling_std(df.Close, 30)
```

```python
df.corr(method='spearman')
```

```python
# Pearson Correlation between total email volume and stock volatility
corr = {}
for i in range(5, 50):
    c = df.Emails.corr(pandas.rolling_std(df.Close, i))
    corr[i] = c
corr = pandas.Series(corr)
corr.plot()
plt.title("Pearson Correlation")
plt.xlabel("Volatility window")
plt.ylabel("Correlation")
```

```python
# Spearman Correlation between total email volume and stock volatility

corr = {}
for i in range(5, 50):
    c = df.Emails.corr(pandas.rolling_std(df.Close, i), method='spearman')
    corr[i] = c
corr = pandas.Series(corr)
corr.plot()
plt.title("Spearman Correlation")
plt.xlabel("Volatility window")
plt.ylabel("Correlation")
```
