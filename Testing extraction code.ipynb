### Randomly select 100 files and test the get_header code.

```python
headers = []
for i in range(100):
    f = random.choice(all_files)
    text = open(f).read()
    headers.append(get_header(text))
```

### Test dictionary writing

```python
import csv
```

```python
f = open("TestOut.csv", "wb")
writer = csv.DictWriter(f, ["Date", "From", "To", "Cc", "Bcc"], dialect='excel-tab')
```

```python
writer.writeheader()
for hed in headers:
    for key, val in hed.items():
        hed[key] = val.strip().replace("\r", "")
    writer.writerow(hed)
    
```
