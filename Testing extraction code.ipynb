# Full Extraction

```python
headers = []
for f in all_files:
    text = open(f).read()
    headers.append(get_header(text))
```

```python
rows = []

for message in headers:
    if "To" in message:
        to = set(message["To"])
    else:
        to = set()
        
    if "Cc" in message:
        to = to.union(message["Cc"])
    if "Bcc" in message:
        to = to.union(message["Bcc"])
    for rcpt in to:
        row = {"Date": message["Date"],
               "MessageID": message["MessageID"],
               "From": message["From"],
               "To": rcpt}
        rows.append(row)
```

```python
f = open("AllEmails.csv", "wb")
writer = csv.DictWriter(f, ["MessageID", "Date", "From", "To"])
writer.writeheader()
for row in rows:
    writer.writerow(row)
```
